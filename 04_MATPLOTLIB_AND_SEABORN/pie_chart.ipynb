```python
import matplotlib.pyplot as plt
```

```python
x =[7,2,3,4,8]

plt.pie(x) 
#basic pie
plt.show()
```

```python
x =[7,2,3,4,8]
activities = ["sleeping","food","work","chores","recreation"]
plt.pie(x, labels=activities) 
#added labels 
#the plotting starts at 0 degree and plots anti-clockwise
plt.show()
```

```python
x =[7,2,3,4,8]
activities = ["sleeping","food","work","chores","recreation"]
plt.pie(x, labels=activities,startangle=90) 
 #the plotting starts at 90 degree and plots anti-clockwise
plt.show()
```

```python
x =[7,2,3,4,8]
activities = ["sleeping","food","work","chores","recreation"]
plt.pie(x, labels = activities,startangle =90,radius =2) # changed the size of pie chart, default =1
plt.show()
```

```python
x =[7,2,3,4,8]
activities = ["sleeping","food","work","chores","recreation"]
plt.pie(x, labels = activities,startangle =90,radius =2, autopct='%1.2f%%') 
# added percentage, formula should have % sign in front and %% at the end.
#the value inside can be 0.1f,0.2f, and so on. 
#Here 2 after decimal indicatas how many decimal points are to be displayed. 
plt.show()
```

```python
x =[7,2,3,4,8]
activities = ["sleeping","food","work","chores","recreation"]
plt.pie(x, labels = activities,startangle =90,radius =2, autopct='%1.2f%%', explode =[0,0,0.2,0,0]) 
# sliced out the important section. 
#Input is a list with number of values =number of activities
#all will have a default value of zero, and the one which needs to explode will have a positive value, typically 0.2
plt.show()
```
